```python
conf = json.loads(configJson)
#print configJson
train,test = loadDataset(configJson) #loads training and test users events, then generates the train and test splits
train.first()
```

```python
trainSessions = train.map(lambda x: genSessions(x, conf, 'train'))
```

```python
trainSessions.filter(lambda x: x[0] == 14336).take(5)
```

```python
trainSessions.filter(lambda x: x[0] == 31240).take(5)
```

```python
# this call generates an RDD having, for each user session with timestamp ts, the list of all the sessions with a timestamp <= ts
# the structure of a row in the RDD is the following:
# (user_id, ts), list_of_sessions
from operator import add
trainSessionHistory = trainSessions.join(trainSessions)\
    .filter(lambda x: x[1][0].ts >= x[1][1].ts)\
    .map(lambda x: ((x[0], x[1][0].ts), [x[1][1]]))\
    .reduceByKey(add)
```

```python
trainSessionHistory.filter(lambda x:x[0][0] == 14336).take(10)
```

```python
# APPLY THE HISTORY AGGREGATION POLICY to the session history RDD
# this will generate a dictionary with the current session and the history
# (represented, for convenience, as a dictionary with the aggregated values for the tracks played in sessions previous to the current one)
trainSessionHistoryAggr = trainSessionHistory.map(lambda x: applyHistoryAggrPolicy(x, conf, find_current=True))
```

```python
trainSessionHistoryAggr.filter(lambda x: x["current"].user_id == 23920).take(10)
```

```python
#APPLY THE TRAINING GENERATION POLICY, then group training lines by user_id
training = trainSessionHistoryAggr.flatMap(lambda x: applyTrainGenPolicy(x, conf))\
    .map(lambda x: (x.user_id, [x]))\
    .reduceByKey(add)
```

```python
training.filter(lambda x: x[0] == 14336).take(1)
```

Generate the test set

```python
test.filter(lambda x: json.loads(x)['linkedinfo']['subjects'][0]['id'] == 16536).take(5)
```

```python
#generate recommendation requests and group them all together for each user
testRequests = test.map(lambda x: genSessions(x, conf, 'test'))\
    .map(lambda x:(x[0], [x[1]]))\
    .reduceByKey(add)
```

```python
testRequests.filter(lambda x: x[0] == 23920).take(1)
```

```python
# generate the history for each group of test requests,
# by taking all sessions from the training set (which are previous to each request by construction)
# NOTE: History is kept constant for all the test requests 
testRequestsHistory = testRequests.join(trainSessions)\
    .map(lambda x: (x[0], (x[1][0],[x[1][1]])))\
    .reduceByKey(lambda x, y: (x[0], x[1]+y[1]))
```

```python
testRequestsHistory.filter(lambda x: x[0] == 23920).take(1)
```

```python
#APPLY THE HISTORY AGGREGATION POLICY TO TEST REQUESTS
testRequestsHistoryAggr = testRequestsHistory.map(lambda x: applyHistoryAggrPolicy(x, conf, find_current=False))
```

```python
testRequestsHistoryAggr.filter(lambda x: x['current'][0].user_id == 23920).take(1)
```

```python
TestEntry = namedtuple('TestEntry', ['req_id', 'user_id', 'history', 'tracks', 'properties'])
testing = testRequestsHistoryAggr.map(lambda x:(x['current'][0].user_id, \
    [TestEntry(req.req_id, req.user_id, x['history'], req.tracks, req.properties) for req in x['current']]))\
    .reduceByKey(add)
```

```python
testing.filter(lambda x: x[0] == 23920).take(2)
```

Apply the collaborativity strategy here

```python
# pick the most recent history previous to the "split timestamp" for each user in trainSessionHistoryAggr
recentTrainHistoryAggr = trainSessionHistoryAggr.map(lambda x: (x['current'].user_id, (x['current'].ts, x['history'])))\
    .filter(lambda x: x[0] < conf['split']['ts'] and x[1][1])\
    .reduceByKey(lambda x, y: x if x[0] > y[0] else y)\
    .map(lambda x: (x[0], x[1][1]))
```

```python
#Generate recommendations in the rigth format, then test
def addResponse(reqJson, respJson):
    req = json.loads(reqJson)
    req['linkedinfo'].update({'response': respJson})
    return json.dumps(req)

responses = test.map(lambda x: (json.loads(x)['id'], x))\
    .join(predictions)\
    .map(lambda x: addResponse(x[1][0], formatResponse(x[1][1], json.loads(x[1][0])['properties']['reclistSize'])))
```
